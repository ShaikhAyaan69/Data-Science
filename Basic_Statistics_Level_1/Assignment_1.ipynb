# **Standarization of Numerical Variables**

```python
'''
Standardization, also known as Z-score normalization,
is nothing but a technique that adjust the data to have a mean of zero and a standard deviation of one.
Its converts different measurements like height in inches and weight in pounds to a common unit so they can be compared easily.
The formula for Z-score is Z=x−μ/σ
where x is data value, μ is mean, σ is standard deviation.
'''
```

```python
# Here we have import zscore for Z-score normalization calculation
from scipy.stats import zscore
```

```python
# Here we have identify numerical columns in dataset using select_dtypes() function for the calculations
numerical_columns = df.select_dtypes(include=['int','float']).columns
numerical_columns
```

```python
# Here we have displayed all numerical columns assigned in 'numerical_columns'
numerical_columns = df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
numerical_columns
```

```python
# Here we have displayed mean of all numerical columns
numerical_columns.mean()
```

```python
# Here we have displayed standard deviation of all numerical columns
numerical_columns.std()
```

Columnwise Z-score calculations

```python
# Here we have created a new column in DataFrame i.e 'Volume Zscore' to store Z-score values of Volumn column
df['Volume Zscore'] = None

# Here we have calculated z-score values of Volume column and stored it in new Volume Zscore column
df['Volume Zscore'] = zscore(df['Volume'])
df['Volume Zscore']
```

# **Conversion of Categorical columns into Dummy Variables**

```python
'''
Conversion of categorical columns into dummy variables i.e 'one-hot encoding', is necessary because many machine learning algorithms
requires numerical inputs and cannot directly handles categorical data like "green", "blue", "red".
One-hot encoding transforms this categorical data into separate binary columns, ensuring data each category is treated independently
without implying any order or hierarchy.
This process helps algorithms to interpret the data correctly, improving model performance and previnting biases.
'''
```

```python
# Here we have identify categorical columns in dataset using select_dtypes() function
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
# Here we have store all categorical columns in new dataframe i.e categorical_columns
categorical_columns = df.select_dtypes(include=['object'])
categorical_columns
```

Applying one-hot encoding and displaying portion of transformed columns

```python
# Here we have applied one-hot encoding to 'Date' column & only 'Date' we have displayed.
categorical_columns_date = pd.get_dummies(categorical_columns['Date'])
categorical_columns_date
```

# **Conclusion**

The descriptive analytics and data visualizations reveal that discounts exhibit a wide range, with a concentration around moderate values and some extreme high and low discounts. There is a noticeable correlation between higher discounts and increased sales volume, though extreme discounts may not always translate to proportional revenue increases. Seasonal trends and promotional events significantly influence discount patterns and sales, with certain product categories responding more effectively to discounts than others. Customer segments also show varied responses to discounts, suggesting that different strategies may be needed for different demographic groups.


Data preprocessing steps like standardization and one-hot encoding are crucial for ensuring effective data analysis and machine learning model performance. Standardization scales numerical features to a uniform range, which helps algorithms that are sensitive to feature scales, such as gradient descent-based methods, to converge more efficiently and accurately. One-hot encoding transforms categorical variables into a numerical format that machine learning algorithms can process, allowing models to correctly interpret and use categorical data. These preprocessing steps enhance model performance by ensuring that all features contribute equally and are in the right format, leading to more reliable and insightful analysis and predictions.
